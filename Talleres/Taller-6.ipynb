```python
datos=genfromtxt("datos.csv", delimiter=",")

longs=datos[:,0] #Estos dos arreglos los usaré mucho más adelante.
lats=datos[:,1]

datos[:,1]=90-datos[:,1] #Acà hallo el complementario de la latitud  y lo reemplazo en la matriz ya que ese es el phi que me interesa.
#Con este reemplazo ya tengo en la primera columna theta, en la segunda phi. El radio de la tierra es el siguiente:
R=6371000

#Inicializo matriz de coordenadas cartesianas.
matrizCart=zeros((len(datos),4), dtype=float)

matrizCart[:,0]=R*sin(radians(datos[:,1]))*cos(radians(datos[:,0])) #Coordenada en X

matrizCart[:,1]=R*sin(radians(datos[:,1]))*sin(radians(datos[:,0])) #Coordenada en Y

matrizCart[:,2]=R*cos(radians(datos[:,1])) #Coordenada en Z

matrizCart[:,3]=((datos[:,3])*3600)+((datos[:,4])*60)+(datos[:,5]) #Tiempo en segundos

#Ahora para inicializar en el punto (0,0,0) y en el tiempo 0, le restaré las 4 componentes a toda la matriz.

X0=matrizCart[0,0]
Y0=matrizCart[0,1]
Z0=matrizCart[0,2]
t0=matrizCart[0,3]


matrizCart[:,0]= matrizCart[:,0]-X0
matrizCart[:,1]= matrizCart[:,1]-Y0
matrizCart[:,2]= matrizCart[:,2]-Z0
matrizCart[:,3]= matrizCart[:,3]-t0

#Esta ya es la matriz normalizada que usaré. 
print(matrizCart)
```

```python
def prodPunto(x1,x2,x3,y1,y2,y3):
    return (x1*y1)+(x2*y2)+(x3*y3)
thetaB=radians(-74.07)
phiB=radians(90-4.5981)
print(phiB)
```

```python
x1,x2,x3=matrizCart[:,0],matrizCart[:,1],matrizCart[:,2]

vecPhi=[-sin(thetaB),cos(thetaB),0]
vecThe=[cos(thetaB)*cos(phiB),sin(thetaB)*cos(phiB),-sin(phiB)]

Bog1=prodPunto(x1,x2,x3,vecPhi[0],vecPhi[1],vecPhi[2])
Bog2=-1*prodPunto(x1,x2,x3,vecThe[0],vecThe[1],vecThe[2])


figure(figsize=(8,8))
plot(Bog1,Bog2, label="Mi paseo por Bogotá")
ylabel("Pos. Vector Norte (m)")
xlabel("Pos. Vector Oriente (m)")
ylim(-2500,1400)
legend()
```

```python
dx=Bog1
dy=Bog2
t=matrizCart[:,3]
D=[]

#En estas dos líneas corremos generamos dos arreglos con la anterior posición de la posicion x y y, respectivamente.
dxx=roll(dx,1)
dyy=roll(dy,1)

#El ejecutar una resta entre los siguientes arreglos me dará el desplazamiento neto en x y y, por cada instante de t.

despX=dx-dxx
despX[0]=0 #Debe empezar en 0 este arreglo.
despY=dy-dyy
despY[0]=0 #Debe empezar en 0 este arreglo.

normDesp=sqrt(despX**2+despY**2) #Obtengo la norma del desplazamiento en cada instante t.


distAcum=cumsum(normDesp)  #El aplicar cumsum me da un arreglo cuyos elementos son la distancia acumulada que ha recorrido el sujeto.



figure(figsize=(8,8))
plot(t,distAcum, label="Distancia Recorrida")
xlabel("Tiempo (s)")
ylabel("Distancia (m)")
legend(loc=4)
```

```python
def darDatos(Dist):
   
    ciclo=True  #Apenas encuentre el dato que nos interesa se saldrá del ciclo.  
    
    for i in range (len(distAcum)):
        
        
        
        if (distAcum[i]>=Dist and ciclo==True):
            T=t[i]
            longi=longs[i]
            lati=lats[i]
            ciclo=False
    
    return(T,longi,lati)


```

```python

for j in range (1,10):
    
    dist=1000*j #Esto hace que lo que se tengan en cuenta para la función de arriba sean metros en múltiplos de 1000. 
    
    Time,longitud,latitud=darDatos(dist)
    print("El tiempo necesario para recorrer", dist, "metros fue", Time, "segundos.")
    print("La coordenadas (long,lat) de este punto son", longitud,"-", latitud)
    print("")
    
print("")
print("Cabe resaltar que estos tiempos son estimados puesto que no se tiene un dato para cada instante.")
```

```python
from scipy import interpolate

#Usaré interpolación para construir datos uniformemente espaciados.
interp=interpolate.interp1d(t, distAcum)
tNuevo=linspace(0,t[len(distAcum)-1],len(distAcum))

y=interp(tNuevo)  #Datos interpolados, uniformemente espaciados.

yforward=roll(y,-1)  #Creo un arreglo con las posiciones de y movidas a la derecha.

h=(tNuevo[-1]-tNuevo[0])/(len(tNuevo)-1)  #Este es el stepsize.

derivadas=(yforward-y)/h   #Derivadas con forward difference.  
derivadas=derivadas[:-1]    #Quito el último dato porque no tiene sentido incluirlo, debido a la anterior resta.

figure(figsize=(9,9))
scatter(tNuevo[:-1],derivadas)  #También remuevo el último dato del tiempo porque en el último instante se asume que no habrá velocidad.
ylim(0,30)
xlim(0,6100)
title("Rapideces durante el paseo en cada instante t")
ylabel("Rapidez (m/s)")
xlabel("Tiempo (s)")
```

Se ve demasiado irregular por la discontinuidad que hay entre los datos, el reemplazar las rapideces que obtuvimos como 0 por el dato que se encuentra inmediatamente antes podría ser una buena solución. El propósito de hacer esto es considerar que el movimiento fue continuo, y que nunca se detuvo por completo. Por otro lado, los cambios tan abruptos en la rapidez del sujeto también son irreales por lo que se implementa un ciclo en el que si el cambio de velocidad es muy drástico se elimina esa velocidad y se reemplaza por un promedio de las rapideces de sus datos inmediatamente cercanos.

```python
hNuevo=h+5

derivadasN=(yforward-y)/(hNuevo)
derivadasN=derivadasN[:-1]

for i in range (len(derivadasN)):
    if (derivadasN[i]==0):
        derivadasN[i]=derivadasN[i-1]
        
for i in range (len(derivadasN)):
    if ((derivadasN[i]-derivadasN[i-1]>3)):
        derivadasN[i]=(derivadasN[i-1]+derivadasN[i+1])/2    
   
        
figure(figsize=(15,9))
scatter(tNuevo[:-1],derivadasN)  
ylim(0,30)
xlim(0,6100)
title("Rapideces durante el paseo en cada instante t")
ylabel("Rapidez (m/s)")
xlabel("Tiempo (s)")
ylim(0,7)
       
        
```

De esta manera, eliminando los datos poco representativos, la gráfica se ve un poco mejor distribuida y tiene mayor continuidad. El disminuir el tamaño del stepsize aumenta la precisión de las medidas. El disminuir los cambios de rapidez (medidos cada segundo) dos ordenes de magnitud hace que los datos sean más válidos y creíbles.

Por último, el que el promedio de las rapideces obtenidas sea cercano a 1 es consistente con la rapidez con la que el sujeto realizó su recorrido. Aproximadamente 1.6 metros por segundo. Aún así, este tan sólo es un primer acercamiento para obtener resultados más razonables, aún se pueden hacer mejorías.
