```python
#Se crea una matriz de ceros, y luego se cambian a 1 los valores solicitados.
Matriz1=zeros((1024,1024), dtype=int)
Matriz1[496:528,496:528]=1

#Se crea la imagen, que tiene dos subplots. La rendija y el patrón de interferencia.
figure(figsize (15,15))
subplot(1,2,1)
imshow(Matriz1, cmap='hot')
title("Ubicación de la rendija")
ylabel("Posición Y")
xlabel("Posición X")

#Se aplica la transformada de fourier, y con la segunda función generamos el patrón de interferencia.
Trans1=fft2(Matriz1)
Patron=fftshift(Trans1)

#Se corta el área solicitada.
Zoom1=Patron[412:612,412:612]
Final1=2*log10(abs(Zoom1)+1)

#Se grafica el patrón de interferencia.
subplot(1,2,2)
imshow(Final1, cmap='gray')
ylabel("Posición Y")
xlabel("Posición X")
title("Patrón de interferencia, Apertura Cuadrada (log resaltado)")
```

```python
#Se crea una nueva matriz de ceros, y se agregan unos en las zonas indicadas. Para crear ka==las dos rendijas
Matriz2=zeros((1024,1024), dtype=int)
Matriz2[412:612,460:480]=1
Matriz2[412:612,564:584]=1

#Lo mismo que en el anterior bloque de código.
figure(figsize(15,15))
subplot(1,2,1)
imshow(Matriz2, cmap='hot')
title("Ubicación de la rendija")
ylabel("Posición Y")
xlabel("Posición X")

#Lo mismo que en el anterior bloque de código.
Trans2=fft2(Matriz2)
Patron2=fftshift(Trans2)

#Cortamos el área solicitada con slice notation. 
Zoom2=Patron2[448:576,448:576]
Final2=(abs(Zoom2)**2) #Se eleva el módulo al cuadrado.

subplot(1,2,2)
imshow(Final2, cmap='hot')
ylabel("Posición Y")
xlabel("Posición X")
title("Patrón de interferencia, Doble Rendija")
```

```python
#Acá recurrimos a la ecuación de un círculo para obtener la rendija circular. 
#El restarle 512 a cada componente es para ubicar el círuclo de radio 10.24 celdas en el centro.

Matriz3=zeros((1024,1024), dtype=int)
i=0
while(i<1024):
    j=0
    while(j<1024):
        if (sqrt((j-512)**2 + (i-512)**2)) < 10.24:
            Matriz3[i,j]=1
        j+=1
    i+=1

#De acá para abajo es lo mismo que en los anteriores bloques.    
figure(figsize(15,15))
subplot(1,2,1)
imshow(Matriz3, cmap='hot')
title("Ubicación de la rendija")
ylabel("Posición Y")
xlabel("Posición X")

Trans3=fft2(Matriz3)
Patron3=fftshift(Trans3)


Final3=(log10(abs(Patron3)**2))
subplot(1,2,2)
imshow(Final3, cmap='BuGn')
ylabel("Posición Y")
xlabel("Posición X")
title("Patrón de interferencia, Apertura Circular")
```

```python
Corte=(abs(Patron3[512,:])**2) #Corte pasando por el centro
IMax=max(Corte)    #Acá obtenemos la amplitud máxima.
CorteFinal=Corte/IMax  #Normalizamos como lo pide el problema.


Coorx=2*pi*0.01*linspace(-512,512,1024)  #Creamos coordenadas en x con el rango que se pide.

figure(figsize(15,7))
scatter(Coorx,CorteFinal,s=7)
xlim(-30,30)
ylim(0,1)
xlabel("2πRj/L", fontsize=15)
ylabel("I / Io", fontsize=15)
```

```python
#Acá incluimos sólo el tramo de 0 a 12 en x como lo pide el enunciado de la tarea.

figure(figsize(15,7))
scatter(Coorx,CorteFinal,s=5)
xlim(0,12)
ylim(0,1)
xlabel("2πRj/L", fontsize=15)
ylabel("I / Io", fontsize=15)
```

```python
#En este bloque se crea la interpolación de todos los puntos del scatter anterior y se vuelve a graficar.

CoorxInterpol=2*pi*0.01*linspace(-512,512,10000000)
Interpolados= interpolate.interp1d(Coorx,CorteFinal, kind='cubic')
figure(figsize(15,7))
plot(CoorxInterpol,Interpolados(CoorxInterpol),c='r')
ylim(0,1)
xlim(0,12)
xlabel("2πRj/L", fontsize=15)
ylabel("I / Io", fontsize=15)
title("Interpolacion de datos para el Patron de Interferencia")
```

```python
#Se realiza el corte es estos puntos, ya que son los más cercanos a 0 y 12. Se tomaron los valores justo antes y justo después.
tramoX=CoorxInterpol[4999999:6865097]
tramoX=linspace(0,12,700000)
interpoladosTramo=Interpolados(tramoX)
#Esto lo hicimos con el fin de utilizar las mismas abscisas y ordenadas de los puntos anteriores.


#La siguiente función fue diseñada para que refrese las coordenadas de los máximos locales de una función que le entra como parámetro.
def darMaximos (f):
    i=0
    resp=[]
    resp2=[]
    while(i<len(f)-1):
        
        if(f[i]>f[i+1] and f[i]>f[i-1]):
            resp.append(f[i])
            resp2.append(tramoX[i])
        i+=1
    return resp,resp2

#La siguiente función fue diseñada para que refrese las coordenadas de los mínimos locales de una función que le entra como parámetro.

def darMinimos (f):
    i=0
    resp=[]
    resp2=[]
    while(i<len(f)-1):
        if(f[i]<f[i+1] and f[i]<f[i-1]):
            resp.append(f[i])
            resp2.append(tramoX[i])
        i+=1
    return resp, resp2

#Ambas funciones devuelven dos arreglos, dónde las coordenadas son Y y X, respectivamente.
```

```python
#Obtenemos los puntos críticos de la interpolación del tramo.


y,x=darMaximos(interpoladosTramo)
y2,x2=darMinimos(interpoladosTramo)
```

```python
#Ploteamos e imprimimos máximos y mínimos locales.
figure(figsize(15,7))
scatter(x,y,c='r',label="Máximos de intensidad")
scatter(x2,y2,label="Mínimos de intensidad")
ylim(-0.02,1.02)
xlim(-0.03,12)
legend()


var=0
print("")
print("Los puntos máximos son: ")
print("")
while(var<len(x)):
   
    print(x[var],"|",y[var])
    var+=1
    
    
var=0 
print("")
print("Los puntos mínimos son: ")
print("")

while(var<len(x2)):
   
    print(x2[var],"|", y2[var])
    var+=1
    
```
