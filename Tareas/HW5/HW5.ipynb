```python
a = 0.24250109
b = -0.0607718825
c = 0.93240737
d = 0.86473146
```

```python
#Definimos una función a la cual se le entrega una rreglo de posiciones y velocidades, y que entrega sus derivadas. 
def Derivada(Y):
    
    #Se hallan las distancias de separacion entre los tres puntos, se elevan al cubo y se invierten, para usarlo como un factor y no un dividendo en la fórmula. 
    Rc12i=(((Y[0]-Y[2])**2)+(Y[1]-Y[3])**2)**(-3/2)
    Rc23i=(((Y[2]-Y[4])**2)+(Y[5]-Y[3])**2)**(-3/2)
    Rc13i=(((Y[0]-Y[4])**2)+(Y[1]-Y[5])**2)**(-3/2)
    
    
    #Se devuelve un arreglo con velocidades y aceleraciones. 
    return array((Y[6],Y[7],Y[8],Y[9],Y[10],Y[11],
                 (Rc12i *(Y[2]-Y[0])+Rc13i *(Y[4]-Y[0])),
                 (Rc12i *(Y[3]-Y[1])+Rc13i *(Y[5]-Y[1])),
                 (Rc12i *(Y[0]-Y[2])+Rc23i *(Y[4]-Y[2])),
                 (Rc12i *(Y[1]-Y[3])+Rc23i *(Y[5]-Y[3])),
                 (Rc13i *(Y[0]-Y[4])+Rc23i *(Y[2]-Y[4])),
                 (Rc13i *(Y[1]-Y[5])+Rc23i *(Y[3]-Y[5]))))
    
    
    #Multiplicamos el inverso de la distancia entre partículas por el vector de director de la fuerza de atracción. La suma nos dará la "atracción total."
```

```python
#Se implementa el código obtenido del notebook de Ecuaciones Diferenciales Ordinarias del Profesor Juan David Lizarazo.
def PasoKutta4():
    
    global Y
    global Deltat
    K1= Derivada(Y)
    K2= Derivada(Y+(Deltat*(1/2*K1)))
    K3= Derivada(Y+(Deltat*(1/2*K2)))
    K4= Derivada(Y+Deltat*(K3))
    Y+=Deltat * (1/6.*K1 + 1/3. * K2 + 1/3.*K3 + 1/6.*K4)
    
#Se ponderan los 4 valores de K para así obtener el valor de los integrandos. 
```

```python

#En la siguiente línea se declara el arreglo Y con las condiciones iniciales provistas por el enunciado. 
Y=array([a,b,-a,-b,0,0,c,d,c,d,-2*c,-2*d])
#En el siguiente arreglo se irán almacenando los resultados.
history=array([[Y[0],Y[1],Y[2],Y[3],Y[4],Y[5],Y[6],Y[7],Y[8],Y[9],Y[10],Y[11]]])
#Se declara el timestep a usar. 
Deltat=1/1000

#Se ejecuta la función de Runge Kutta orden 4, hasta lograr alcanzar la media unidad de tiempo. 
t=0
while(t<0.5):
    PasoKutta4()
    history=append(history,[[Y[0],Y[1],Y[2],Y[3],Y[4],Y[5],Y[6],Y[7],Y[8],Y[9],Y[10],Y[11]]],axis=0)
    t+=Deltat

    
#Se grafican las trayectorias de las tres partículas usando subplot.     
figure(figsize(20,7))
subplot(1,3,1)
plot(history[:,0],history[:,1],label="Trayectoria")
scatter(history[0,0],history[0,1],c='r',s=70,label="Inicio")
scatter(history[500,0],history[500,1],c='g',s=70,label="Fin")
legend(scatterpoints=1)
xlabel("Posición x")
ylabel("Posición Y")
ylim(-0.3,0.3)
title("Partícula 1")

subplot(1,3,2)
plot(history[:,2],history[:,3],label="Trayectoria")
scatter(history[0,2],history[0,3],c='r',s=70,label="Inicio")
scatter(history[500,2],history[500,3],c='g',s=70,label="Fin")
legend(scatterpoints=1)
xlabel("Posición x")
ylabel("Posición Y")
ylim(-0.3,0.3)
title("Partícula 2")

subplot(1,3,3)
plot(history[:,4],history[:,5],label="Trayectoria")
scatter(history[0,4],history[0,5],c='r',s=70,label="Inicio")
scatter(history[500,4],history[500,5],c='g',s=70,label="Fin")
legend(scatterpoints=1)
xlabel("Posición x")
ylabel("Posición Y")
ylim(-0.3,0.3)
title("Partícula 3")
```

Se resuelve la ecuación diferencial con el mismo proceso que en el ejercicio anterior. Se enumeran los indices usado a continuación para tener una mayor claridad.

```python
#0=x1
#1=y1
#2=x2
#3=y2
#4=x3
#5=y3
#6=x4
#7=y4
#8=vx1
#9=vy1
#10=vx2
#11=vy2
#12=vx3
#13=vy3
#14=vx4
#15=vy4
def Derivada2(Y):
    Rc12i=(((Y[0]-Y[2])**2)+(Y[1]-Y[3])**2)**(-3/2)
    Rc13i=(((Y[0]-Y[4])**2)+(Y[1]-Y[5])**2)**(-3/2)
    Rc14i=(((Y[0]-Y[6])**2)+(Y[1]-Y[7])**2)**(-3/2)    
    Rc23i=(((Y[2]-Y[4])**2)+(Y[5]-Y[3])**2)**(-3/2)
    Rc24i=(((Y[2]-Y[6])**2)+(Y[3]-Y[7])**2)**(-3/2)
    Rc34i=(((Y[4]-Y[6])**2)+(Y[5]-Y[7])**2)**(-3/2)

    return array((Y[8],Y[9],Y[10],Y[11],Y[12],Y[13],Y[14],Y[15],
                 (Rc12i *(Y[2]-Y[0])+Rc13i *(Y[4]-Y[0])+Rc14i *(Y[6]-Y[0])),
                 (Rc12i *(Y[3]-Y[1])+Rc13i *(Y[5]-Y[1])+Rc14i *(Y[7]-Y[1])),
                 (Rc12i *(Y[0]-Y[2])+Rc23i *(Y[4]-Y[2])+Rc24i *(Y[6]-Y[2])),
                 (Rc12i *(Y[1]-Y[3])+Rc23i *(Y[5]-Y[3])+Rc24i *(Y[7]-Y[3])),
                 (Rc13i *(Y[0]-Y[4])+Rc23i *(Y[2]-Y[4])+Rc34i *(Y[6]-Y[4])),
                 (Rc13i *(Y[1]-Y[5])+Rc23i *(Y[3]-Y[5])+Rc34i *(Y[7]-Y[5])),
                 (Rc14i *(Y[0]-Y[6])+Rc24i *(Y[2]-Y[6])+Rc34i *(Y[4]-Y[6])),
                 (Rc14i *(Y[1]-Y[7])+Rc24i *(Y[3]-Y[7])+Rc34i *(Y[5]-Y[7]))))

    
```

```python
#Definimos la función que integra la ecuación diferencial nuevamente.
def PasoKutta4():
    global Y
    global Deltat
    K1= Derivada2(Y)
    K2= Derivada2(Y+(Deltat*(1/2*K1)))
    K3= Derivada2(Y+(Deltat*(1/2*K2)))
    K4= Derivada2(Y+Deltat*(K3))
    Y+=Deltat * (1/6*K1 + 1/3 * K2 + 1/3*K3 + 1/6*K4)
```

```python
#Se declaran las condiciones iniciales.
b=0.384277200514
c=0.0
d=-0.0156823005697
e=-0.13966430504
f=-0.352912599375
g=0.0
h=-0.0156823005697
i=0.13966430504
j=0.0

#inicializamos a en 0.9, para corroborar que la ecuación diferencial esté funcionando correctamente
a=0.9

k=-2.01155925929
l=-1.19817066623
m=0.0
n=1.63619158614
o=2.01155925929
p=-1.19817066623
```

```python
#Se asignan los valores de condiciones iniciales y se itera sobre una unidad de tiempo la integración.
Y=array([b,c,d,e,f,g,h,i,j,a,k,l,m,n,o,p])
history=array([[Y[0],Y[1],Y[2],Y[3],Y[4],Y[5],Y[6],Y[7],Y[8],Y[9],Y[10],Y[11],Y[12],Y[13],Y[14],Y[15]]])
Deltat=3.33*10**(-4)

t=0
while(t<1):
    PasoKutta4()
    history=append(history,[[Y[0],Y[1],Y[2],Y[3],Y[4],Y[5],Y[6],Y[7],Y[8],Y[9],Y[10],Y[11],Y[12],Y[13],Y[14],Y[15]]],axis=0)
    t+=Deltat
```

```python
Deltat=3.33*10**(-4) #Timestep propuesto por el enunciado.

#Definimos una función, a la cual se le entrega un valor para A, y este devuelve el radio entre la posición inicial y final de un "ciclo"
def Minimo(A):
    global Y
    global history
    Y=array([b,c,d,e,f,g,h,i,j,A,k,l,m,n,o,p])
    history=array([[Y[0],Y[1],Y[2],Y[3],Y[4],Y[5],Y[6],Y[7],Y[8],Y[9],Y[10],Y[11],Y[12],Y[13],Y[14],Y[15]]])
    

    #Se ejecuta el ciclo durante una unidad de tiempo.
    t=0
    while(t<1):
        PasoKutta4()
        history=append(history,[[Y[0],Y[1],Y[2],Y[3],Y[4],Y[5],Y[6],Y[7],Y[8],Y[9],Y[10],Y[11],Y[12],Y[13],Y[14],Y[15]]],axis=0)
        t+=Deltat
    
    xf=history[-1,0]
    yf=history[-1,1]
    
    desfase=((xf-b)**2+(yf-c)**2)
    
    return(desfase)
    
```

```python
#Ejecutamos la función anterior entre 0.7 y 0.78 para la primera partícula.
Minimos=[]
ii=0.7
while(ii<=0.78):
    
    Minimos.append(Minimo(ii))
    ii+=0.001
    
x=linspace(0.7,0.78,len(Minimos))
minarray=np.array(Minimos)

figure(figsize(10,10))
plot(x,minarray)
ylabel("Diferencia de radios")
xlabel("Valor de a")
title("Diferencia de radios para la partícula 1 en función de a")
xlim(0.69,0.79)

min(Minimos)
```

```python
#Se define el nuevo rango de búsqueda. Y se vuelven a buscar los valores de separación.
lim1=0.76
lim2=0.761

Minimos=[]
ii=lim1
while(ii<=lim2):
    
    Minimos.append(Minimo(ii))
    ii+=0.0001
    
x=linspace(lim1,lim2,len(Minimos))

figure(figsize(10,10))
plot(x,Minimos)
xlim(0.755,0.765)
ylabel("Diferencia de radios")
xlabel("Valor de a")
title("Diferencia de radios para la partícula 2 en función de a (Rango menor)")

min(Minimos)
```

Es sabido que una si una función tiene un valor mínimo en un punto, su derivada cambia de signo en ese mismo punto. Por esta razón, derivamos numéricamente la función encontramos el punto de corte de esta, y así conoceríamos a.

```python
H = (lim2-lim1)/len(Minimos) # Este es el timestep para el cual el método de Central Difference es más preciso. Tomado de las presentaciones de clase.
yforward = np.roll(Minimos,-1) # toma el arreglo y y lo mueve una posición hacia la izquierda
ybackward = np.roll(Minimos,1) # se mueve el arreglo una posición a la derecha.


dy = (yforward-ybackward)/(2.*H)
#Se remueven la primera y última posición porque no tienen sentido.
dy=dy[1:-1]


#Hacemos lo correspondiente con x para que tengan la misma dimensión.
x=x[1:-1]

#graficamos los datos de derivada y un ajuste lineal de esta.
figure(figsize(10,10))
scatter(x,dy,label="Datos reales")
title("Derivada de distancia de separación en función de a")
xlabel("Valor de a")
ylabel("Derivada")
xlim(0.759,0.762)
#Hallamos el mayor ajuste.
pen,cor=polyfit(x, dy, deg=1)
plot(x,pen*x+cor,label="Ajuste lineal",c='r')
legend()
```

```python
#Usamos fsolve de scipy, para encontrar el cero de la función lineal.
def lineal(x):
    return pen*x+cor
afinal=fsolve(lineal,1)
print("El valor de a necesario para que las trayectorias sean periodícas es:","%.5f" % afinal)
```

```python
#Definimos una función de energía cinética, usando los valores de velocidad de las componentes al cuadrado.
def Cinetica(i):
    total=(history[i,8]**2+history[i,9]**2+
    history[i,10]**2+history[i,11]**2+
    history[i,12]**2+history[i,13]**2+
    history[i,14]**2+history[i,15]**2)/2
    return total

#Definimos una función de energía potencial, usando el inverso del radio entre las partículas.
def Potencial(i):
    Rc12i=(((history[i,0]-history[i,2])**2)+((history[i,1]-history[i,3])**2))**(-1/2)
    Rc13i=(((history[i,0]-history[i,4])**2)+(history[i,1]-history[i,5])**2)**(-1/2)
    Rc14i=(((history[i,0]-history[i,6])**2)+(history[i,1]-history[i,7])**2)**(-1/2)
    Rc23i=(((history[i,2]-history[i,4])**2)+(history[i,5]-history[i,3])**2)**(-1/2)
    Rc24i=(((history[i,2]-history[i,6])**2)+(history[i,3]-history[i,7])**2)**(-1/2)
    Rc34i=(((history[i,4]-history[i,6])**2)+(history[i,5]-history[i,7])**2)**(-1/2)
    total=-Rc12i-Rc13i-Rc14i-Rc23i-Rc24i-Rc34i
    return total
```

```python
#Ejecutamos las funciones anteriores para todos los valores de tiempo.

#Se declaran los 4 arreglos que almacenan, las 3 energías y el tiempo.
cineticas=[] 
potenciales=[]
total=[]
time=[]
i=0
while(i<len(history)):
    cineticas.append(Cinetica(i))
    potenciales.append(Potencial(i))
    time.append(Deltat*i)
    i+=1
cineticas=array(cineticas)
potenciales=array(potenciales)
total=cineticas+potenciales
```
