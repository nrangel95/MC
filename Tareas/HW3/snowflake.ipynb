```python
def partirLineas(x1,y1,x2,y2):
    xresp=[]
    xresp.append(x1)
    yresp=[]
    yresp.append(y1)
    angulo=arctan((y2-y1)/(x2-x1))
    print((y2-y1)/(x2-x1))
    longitud=sqrt((x2-x1)**2+(y2-y1)**2)
    longitudd=sqrt(13)*longitud/6
    anguloo=angulo-arctan(2/3)
    a1=(longitud/3)*cos(abs(angulo))+x1
    xresp.append(a1)
    b1=(longitud/3)*sin(angulo)+y1
    yresp.append(b1)
    if anguloo>angulo:
        a3=(longitudd)*cos(anguloo)-x1
    else:
        a3=(longitudd)*cos(anguloo)+x1
    xresp.append(a3)
    if anguloo>angulo:
        b3=(longitudd)*sin(anguloo+pi)-y1
    else:
        b3=(longitudd)*sin(anguloo)+y1
    yresp.append(b3)
    a2=(2*longitud/3)*cos(angulo)+x1
    xresp.append(a2)
    b2=(2*longitud/3)*sin(angulo)+y1
    yresp.append(b2)
    xresp.append(x2)
    yresp.append(y2)
    return xresp, yresp

    
```

```python
def rotar(err1,err2,theta):
    temp1=[]
    temp2=[]
    temp3=[]
    temp4=[]
    temp5=[]
    temp6=[]
    j=1
    x=err1[len(err1)-1]-err1[0]
    y=err2[len(err2)-1]-err2[0]
    long=sqrt(x**2+y**2)
    while(j<len(err1)):
        ang=arctan(err2[j]/(err1[j]))
        longch=sqrt(err2[j]**2+err1[j]**2)
        puntox=longch*cos(ang-pi/3)
        puntoy=longch*sin(ang-pi/3)
        temp1.append(puntox+long)
        temp2.append(puntoy)
        j+=1
    j=1
    while(j<len(err1)-1):
        puntox=(temp1[j-1]*(-1))
        puntoy=temp2[j-1]
        temp3.append(puntox+3*long)
        temp4.append(puntoy)
        j+=1
    
    j=0   
    while(j<len(err1)):
        temp5.append(err1[j]+2*long)
        temp6.append(err2[j])
        j+=1
    return temp1,temp2,temp3,temp4,temp5,temp6
```

```python
%pylab inline
import turtle

#Inicializamos la tortuga que usaremos, la pantalla y sus colores.
RR=turtle.Turtle()
pantalla = turtle.Screen()
RR.color("blue")
pantalla.bgcolor("white")

def koch(t, paso, largo):
    
    if paso==0:                 #Cuando el paso es cero, la tortuga escribirá una línea recta de "largo" pixeles.
        t.forward(largo)             
    
    else:
        koch(t, paso-1, largo/3)       #La tortuga escribe un tercio del largo como parámetro. 
        t.left(60)                 #se voltea 60 grados, porque es un triángulo equilatero.
        koch(t, paso-1, largo/3)       #La tortuga vuelve a escribir un tercio del largo como parámetro en la nueva dirección. 
        t.right(120)    
        koch(t, paso-1, largo/3)      #""
        t.left(60)    
        koch(t, paso-1, largo/3)       #Acá ya ha dibujado una línea completa del fractal.
        

pantalla.exitonclick()

koch(RR,0,500)
```
