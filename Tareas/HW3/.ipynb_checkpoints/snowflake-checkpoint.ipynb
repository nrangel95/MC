```python
import turtle

turtle.setup(700, 700) #Tamaño de 500x500 piexeles para la pantalla
pantalla=turtle.Screen() #Inicializamos la pantalla.
pantalla.bgcolor("white") 


#Inicializo 3 tortugas.
RR=turtle.Turtle() 
RR.speed(2000) 
RR2 = turtle.Turtle() 
RR2.speed(2000) 
RR3 = turtle.Turtle() 
RR3.speed(2000)

#Escondo las tortugas
RR.hideturtle()
RR2.hideturtle()
RR3.hideturtle()

#Ubico tortuga1
RR.penup()
RR.left(135)
RR.forward(150)
RR.right(135)
RR.pendown()

#Ubico mi tortuga 2 en el vértice 2
RR2.penup() 
RR2.left(135)
RR2.forward(150)
RR2.right(135)
RR2.forward(400)
RR2.right(120)
RR2.pendown()

#Ubico mi tortuga 3 en el vértice inferior
RR3.penup()
RR3.left(135)
RR3.forward(150)
RR3.right(135)
RR3.right(60)
RR3.pendown()




size = 30 #Tamaño de la figura


#A grandes rasgos lo que el metodo que escribimos hará es generar tres líneas de largo L del fractal de koch desde los tres vértices de un triángulo 
#largo tambiés es L. La tortuga 3 es la única que tiene los ángulos invertidos ya que necesitamos que el copo se haga hacia afuera.
def copo(t1, t2, t3, orden, l): #Funcón copo recibe como parámetro el orden y largo del fractal.
    
    if(orden==0):        #Se crea tres líneas derechas
        t1.forward(l)
        t2.forward(l)
        t3.forward(l)
    
    else:
        copo(t1,t2,t3, orden-1, l/3)       #La tortuga escribe un tercio del largo como parámetro. 
        t1.left(60)#se voltea 60 grados, porque es un triángulo equilatero.
        t2.left(60)
        t3.right(60)
        
        copo(t1,t2,t3, orden-1, l/3)       #La tortuga vuelve a escribir un tercio del largo como parámetro en la nueva dirección. 
        t1.right(120) 
        t2.right(120)
        t3.left(120)
        
        copo(t1,t2,t3, orden-1, l/3)      #""
        t1.left(60)
        t2.left(60)
        t3.right(60)
        
        
        copo(t1,t2,t3, orden-1, l/3)       #Acá ya ha dibujado una línea completa del fractal.
        
        

copo(RR, RR2, RR3, 0, 400) #Crear snowflake de largo 200 con pasos como parámetro.
#En el segundo parámetro, se puede introducir el paso deseado.

pantalla.exitonclick() #Se declara la opción de cerrar la ventana.
```
